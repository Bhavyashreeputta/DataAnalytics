```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#importing the data
df = pd.read_csv("cars.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#1.mpg over the years
plt.bar(df.year, df.mpg, color = 'red')
plt.xlabel("Years")
plt.ylabel("mpg")
plt.show()
```

From the above graph it is clear that there's an improvement in miles per gallon(mpg) over the years but its not constant.Theres increment and decrement over the years in mpg. Year 1981 has the highes mpg.

```python
#2.cubic inches vs time to reach 60 miles per hour
sns.set(rc={'figure.figsize':(12,20)})
sns.barplot(x = "time_to_60", y = "cubicinches", data = df, ci = None, orient = 'h')
```

From the above graph it is clear that cubicinches and time to reach 60 are inversely propotional i.e. higher the volume lesser the time taken to reach 60 miles.

```python
#3 country which prefers vehicles with more horsepower
#considering mean of hp
cars = df[df["hp"]>=106] 
print(cars)
```

```python
cars.reset_index(drop=True, inplace=True)
cars
```

```python
print(df.brand.value_counts())
```

```python
print(cars.brand.value_counts())
plt.pie(cars.brand.value_counts(), labels = ["US", "Japan", "Europe"], shadow = True, autopct='%1.1f%%', startangle = 360)
plt.show()
```

From the above analysis by considering mean of the hp: 1.U.S has manufactured 87.8% vehicles which has horsepower(hp) greater than mean(86 out of 162). 2.Europe has manufactured 5.1% vehicles which has horsepower(hp) greater than mean(51 out of 5).   3.Japan has manufactured 7.1% vehicles which has horsepower(hp) greater than mean(7 out of 48).

Therefore we can conclude that U.S. prefers vehicles with more horsepower

```python
#4.effect of horsepower on the average of the vehicle(mpg)
bins = [0,20,40,60,80,100,120,140,160,180,200,220,240,260]
plt.hist(df.hp, bins, histtype ='bar', rwidth = 0.8)
plt.xlabel("Horsepower")
plt.ylabel("Average of vehicle")
plt.show()
```

```python
#5.cubicinches effect on average of vehicle
sns.set(rc={'figure.figsize':(15,10)})
ax = sns.lineplot(x = 'cubicinches', y = 'mpg', data = df, ci = False)
```

```python
#6.effect of no. of cylinders on mpg
sns.set(rc={'figure.figsize':(7,7)})
plt.xlabel("No. of cylinders")
plt.ylabel("Miles per Gallon")
plt.bar(df.cylinders, df.mpg)
plt.show()
```

```python
#7.Relationship between cubicinches, no.of cylinders, hp
x = df.cylinders
width = 0.4
plt.bar(x, df.hp, width, color = 'blue', label = 'hp')
plt.bar(x+width, df.cubicinches, width, color='green', label = 'cubicinches')
plt.xlabel("cylinders")
plt.xticks(x+width/2, df.cylinders)
plt.legend(loc = "best")
plt.show()
```

From the given data :

case 1: cylinders v/s hp
increase in number of cylinders results in increase in horsepower most of the time.

case 2: cylinders v/s cubicinches
increase in numbers of cylinders resu;ts in increase of cubicinches
